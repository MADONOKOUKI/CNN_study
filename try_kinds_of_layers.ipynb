```python
image_list = []
label_list = []
image_list_cross = []
for dir in os.listdir("data/train"):
  if dir == ".DS_Store":
    continue

  dir1 = "data/train/" + dir
  label = 0

  if dir == "orion": #onionのlabel = 0
    label = 0
  elif dir == "carot": #carotのlabel = 0
    label = 1
  elif dir == "potato":
    label = 2
  elif dir == "CurryPowder":
    label = 3
  else:
    label = 4
  for file in os.listdir(dir1):
    if file != ".DS_Store":
      # 配列label_listに正解ラベルを追加(りんご:0 オレンジ:1)
      label_list.append([label])
      filepath = dir1 + "/" + file
      # 画像を25x25pixelに変換し、1要素が[R,G,B]3要素を含む配列の25x25の２次元配列として読み込む。
      # [R,G,B]はそれぞれが0-255の配列。
      image = np.array(Image.open(filepath).resize((25,25)))
      print(filepath)
      # 配列を変換し、[[Redの配列],[Greenの配列],[Blueの配列]] のような形にする。
      print(image.shape)
#       image = image.transpose(2,0,1)
      #さらにフラットな1次元配列に変換、最初の1/3がred,次がgreen、最後がblueの要素がフラットに並ぶ
      print(image.shape)
      image_list.append(image/255.)
      image = image.reshape(1,image.shape[0] * image.shape[1] * image.shape[2]).astype("float32")[0]
      #配列をimage_listに保存する
      print(image.shape)
      image_list_cross.append(image/255.)

# kerasに渡すためにnumpy配列に変換
image_list = np.array(image_list)
image_list_cross = np.array(image_list_cross)
print(image_list_cross.shape)
# 0 -> [1,0], 1 -> [0,1] という感じ。
#ラベルの配列を1と0からなるラベル配列に変換する

from keras.utils import np_utils # utilities for one-hot encoding of ground truth values
num_classes = 5
print(num_classes)
# print(label_listdtype)
label_list = np.array(label_list).astype('uint8')
print(label_list)
label_list = np_utils.to_categorical(label_list, num_classes) # One-hot encode the labels
# print(num_classes)
Y = to_categorical(np.array(label_list))
# print(Y.shape)
# model = multi_layer_perceptron()
# #optimizer → Adamを使用
# opt = Adam(lr=0.001)
# #compiling model
# model.compile(loss="categorical_crossentropy",optimizer=opt,metrics=["accuracy"])
# #学習の実行
# # epochs 繰り返し学習させる回数
# model.fit(image_list,Y,nb_epoch=1500,batch_size = 100,validation_split=0.1)
print("finish")
```

```python
batch_size = 32 # in each iteration, we consider 32 training examples at once
num_epochs = 200 # we iterate 200 times over the entire training set
kernel_size = 3 # we will use 3x3 kernels throughout
pool_size = 2 # we will use 2x2 pooling throughout
conv_depth_1 = 32 # we will initially have 32 kernels per conv. layer...
conv_depth_2 = 64 # ...switching to 64 after the first pooling layer
drop_prob_1 = 0.25 # dropout after pooling with probability 0.25
drop_prob_2 = 0.5 # dropout in the FC layer with probability 0.5
hidden_size = 512 # the FC layer will have 512 neurons

num_train, height, width, depth = image_list.shape
inp = Input(shape=(height, width, depth)) # depth goes last in TensorFlow back-end (first in Theano)# Conv [32] -> Conv [32] -> Pool (with dropout on the pooling layer)
conv_1 = Convolution2D(conv_depth_1, (kernel_size, kernel_size), padding='same', activation='relu')(inp)
conv_2 = Convolution2D(conv_depth_1, (kernel_size, kernel_size), padding='same', activation='relu')(conv_1)
pool_1 = MaxPooling2D(pool_size=(pool_size, pool_size))(conv_2)
drop_1 = Dropout(drop_prob_1)(pool_1)
# Conv [64] -> Conv [64] -> Pool (with dropout on the pooling layer)
conv_3 = Convolution2D(conv_depth_2, (kernel_size, kernel_size), padding='same', activation='relu')(drop_1)
conv_4 = Convolution2D(conv_depth_2, (kernel_size, kernel_size), padding='same', activation='relu')(conv_3)
pool_2 = MaxPooling2D(pool_size=(pool_size, pool_size))(conv_4)
drop_2 = Dropout(drop_prob_1)(pool_2)
# Now flatten to 1D, apply FC -> ReLU (with dropout) -> softmax
flat = Flatten()(drop_2)
hidden = Dense(hidden_size, activation='relu')(flat)
drop_3 = Dropout(drop_prob_2)(hidden)
out = Dense(num_classes, activation='softmax')(drop_3)
model = Model(inputs=inp, outputs=out) # To define a model, just specify its input and output layers

model.compile(loss='categorical_crossentropy', # using the cross-entropy loss function
              optimizer='adam', # using the Adam optimiser
              metrics=['accuracy']) # reporting the accuracy

# model.fit(image_list, label_list,                # Train the model using the training set...
#           batch_size=batch_size, epochs=30,
#           verbose=1, validation_split=0.1) # ...holding out 10% of the data for validation
# model.evaluate(image_list[25:50], label_list[25:50], verbose=1)  # Evaluate the trained model on the test set!
def train_and_evaluate_model(model,data,labels,data_test,labels_test):
# def train_and_evaluate_model(model, data[train], labels[train], data[test], labels[test]):
    model.fit(data, labels, batch_size=batch_size, epochs=30,verbose=1, validation_split=0.1) # ...holding out 10% of the data for validation
    return model.evaluate(data_test, labels_test, verbose=1)  # Evaluate the trained model on the test set!
#損失値と評価の値を算出している
# from sklearn.model_selection import cross_val_score
# from sklearn.model_selection import StratifiedKFold
# kfold = StratifiedKFold(n_splits=2, shuffle=True, random_state=1)
# print(cross_val_score(model, image_list, label_list,cv=kfold)).mean()  
from sklearn.cross_validation import StratifiedKFold
label = np.r_[np.repeat(0,55), np.repeat(1,1)]
skf = StratifiedKFold(label, n_folds=54, shuffle=True)
total_score = 0
rep = 0
print("success")
for i, (train, test) in enumerate(skf):
    print("TRAIN:", train, "TEST:", test)
    total_score += train_and_evaluate_model(model, image_list[train], label_list[train], image_list[test], label_list[test])[1]
    rep+=1
print(total_score/rep)
# from keras.wrappers.scikit_learn import KerasClassifier
# from sklearn.model_selection import StratifiedKFold
# from sklearn.model_selection import cross_val_score

# print(image_list_cross.shape)
# print(label_list.shape)

# # print(image_list[0],image_list[1]*image_list[2]*image_list[3])
# # image_list = image_list.reshape([56,image_list[1]*image_list[2]*image_list[3],1,1]).astype("float32")

# # for i in range(0,len(image_list)):
# #     image_list_fix[i] = image_list.reshape(image_list[i][0],image_list[i][1]*image_list[i][2]*image_list[i][3]).astype("float32")

# # print(image_list_fix)

# # KerasClassifier/KerasRegressor can be used as same as scikit_learn estimator.
# model = KerasClassifier(build_fn=model, epochs=10, batch_size=10)
# print(label_list)
# # scikit_learn's cross-validation flow
# kfold = StratifiedKFold(n_splits=2, shuffle=True, random_state=1)
# image_list_cross_fix = []
# label_list_cross_fix = []
# for i in range(0,len(image_list_cross)):
#     image_list_cross_fix.append(image_list_cross[i][0]*image_list_cross[i][1])
# print( image_list_cross_fix)
# for i in range(0,len(label_list)):
#     val = 1
#     for j in range(0,len(label_list[i])):
#         val *= label_list[i][j]
#     label_list_cross_fix.append(val)
# print(label_list_cross_fix)
# # print(np.array(image_list_cross_fix))
# # print(np.array(label_list_cross_fix))
# # image_list_cross = image_list_cross.reshape(image_list_cross[0],)
# # label_list = label_list.reshape(label_list[0]*label_list[1])
# # image_list_cross_fix = np.array(image_list_cross_fix)
# # label_list_cross_fix = np.array(label_list_cross_fix)
# # image_list_cross_fix = image_list_cross_fix.resize(56,)
# # label_list_cross_fix = label_list_cross_fix.resize(56,)
# # print(np.array(image_list_cross_fix))
# # print(np.array(label_list_cross_fix))
# results = cross_val_score(model, image_list_cross_fix,label_list_cross_fix, cv=kfold)
# print(results.mean())
```
